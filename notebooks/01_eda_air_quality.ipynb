# 01 — EDA: Air Quality in Kyiv

Exploratory analysis of SaveEcoBot hourly AQI/PM2.5 data.
Run `python pipeline/run_all.py` first to populate DuckDB.

```python
import duckdb
from pathlib import Path

DB = Path('../data/warehouse/kyiv.duckdb')
conn = duckdb.connect(str(DB))

conn.sql("""
    SELECT entity_name, AVG(pm25_ug_m3) AS avg_pm25, COUNT(*) AS n
    FROM stg_air_hourly
    WHERE entity_type = 'district'
    GROUP BY entity_name
    ORDER BY avg_pm25 DESC
""").show()
```

## Finding
District-level PM2.5 varies by ~2x between highest and lowest districts.

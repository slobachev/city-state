# 03 — Mobility Proxy Analysis

Compare GTFS stop density with district PM2.5 averages.

```python
import duckdb
from pathlib import Path

conn = duckdb.connect(str(Path('../data/warehouse/kyiv.duckdb')))
conn.sql("""
    SELECT d.district_name, d.pm25_avg, m.mobility_stop_density
    FROM mart_district_summary d
    JOIN mart_mobility_proxy m ON d.district_slug = m.district_slug
    ORDER BY d.pm25_avg DESC
""").show()
```
